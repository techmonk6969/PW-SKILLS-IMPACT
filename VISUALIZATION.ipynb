Seaborn is a powerful Python data visualization library based on Matplotlib. Here are five types of plots that you can create using Seaborn, along with their common use cases:

Scatter Plot:

Use Case: Scatter plots are used to visualize the relationship between two continuous variables. They are helpful in identifying patterns, clusters, or trends in the data.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='variable1', y='variable2', data=df)
plt.show()
```

Line Plot:

Use Case: Line plots are effective for displaying trends over time. They are commonly used to show how a variable changes with respect to another variable.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.lineplot(x='time', y='value', data=df)
plt.show()
```

Bar Plot:

Use Case: Bar plots are useful for comparing the values of different categories. They are commonly used to show the distribution of a categorical variable.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='category', y='value', data=df)
plt.show()
```

Histogram:

Use Case: Histograms are used to visualize the distribution of a single variable. They help in understanding the underlying frequency or probability distribution.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(x='variable', data=df, bins=30)
plt.show()
```

Heatmap:

Use Case: Heatmaps are effective for visualizing the correlation between variables in a dataset. They are commonly used in exploratory data analysis to identify patterns and relationships.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(data=df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot using Seaborn
sns.lineplot(x='timepoint', y='signal', hue='event', style='region', data=fmri_data)

# Set plot labels and title
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.title('Line Plot of fmri Data')

# Show the plot
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Create two box plots using Seaborn
plt.figure(figsize=(12, 6))

# Box plot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Box Plot of Age by Pclass')

# Box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Box Plot of Fare by Pclass')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds_data = sns.load_dataset("diamonds")

# Create a histogram with hue for 'cut'
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple='stack', bins=30, kde=True)

# Set plot labels and title
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Histogram of Price with Cut Hue')

# Show the plot
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix suitable for a heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, cmap='Blues', annot=True, fmt='d', linewidths=.5)

# Set plot labels and title
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Passenger Count Heatmap for Flights')

# Show the plot
plt.show()
```
