Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

**Matplotlib:**
Matplotlib is a 2D plotting library for Python that enables the creation of high-quality charts, plots, and figures. It provides a wide variety of functionalities for visualizing data in a clear and effective manner. Matplotlib is widely used for data visualization in fields such as data science, machine learning, scientific research, and more.

**Why is it used?**
Matplotlib is used for various purposes, including:

1. **Data Visualization:** Matplotlib allows users to create a wide range of plots and charts to visualize data, making it easier to interpret and analyze.

2. **Exploratory Data Analysis (EDA):** It is a powerful tool for exploring and understanding the underlying patterns, trends, and relationships in the data.

3. **Publication-Quality Plots:** Matplotlib provides fine-grained control over plot appearance, allowing users to create publication-quality plots for presentations, reports, and publications.

4. **Integration with Jupyter Notebooks:** Matplotlib is commonly used in Jupyter Notebooks for interactive data exploration and visualization.

5. **Customization:** It offers a high level of customization, allowing users to control almost every aspect of the plot, including colors, markers, labels, legends, and more.

**Five plots that can be plotted using the Pyplot module of Matplotlib:**

1. **Line Plot:** Used to represent data points with lines connecting them. `plt.plot()`

2. **Scatter Plot:** Displays individual data points without connecting lines, useful for visualizing relationships between two variables. `plt.scatter()`

3. **Bar Plot:** Represents data using rectangular bars. It is suitable for comparing categorical data. `plt.bar()`

4. **Histogram:** Visualizes the distribution of a dataset by dividing it into bins and displaying the frequency of each bin. `plt.hist()`

5. **Pie Chart:** Displays data in a circular statistical graphic divided into slices to illustrate numerical proportions. `plt.pie()`

These are just a few examples, and Matplotlib provides many other plot types and customization options to cater to various visualization needs.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(3)

# Generate data for x and y
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Plot scatter plot
plt.scatter(x, y)

# Add title, xlabel, and ylabel
plt.title('Scatter Plot')
plt.xlabel('x axis')
plt.ylabel('y axis')

# Display the plot
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line Plot 4')

# Adjust layout to prevent clipping of titles
plt.tight_layout()

# Display the plots
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

Bar Plot:
A bar plot is a graphical representation of categorical data where individual bars or columns represent different categories. The length or height of each bar corresponds to the quantity or value associated with that category. Bar plots are used to visually compare and display the values of different categories or groups.

Why is it used?
Bar plots are used for several reasons:

Comparison: Bar plots allow easy comparison of values among different categories or groups.

Categorical Data: They are suitable for displaying categorical data, where each bar represents a category.

Visibility: They provide a clear and visual representation of the relative sizes or quantities of different categories.

Presentation: Bar plots are commonly used in presentations, reports, and publications to convey information effectively.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar Plot
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.title('Bar Plot - Company Profits')
plt.show()

# Horizontal Bar Plot
plt.barh(company, profit, color='salmon')
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')
plt.title('Horizontal Bar Plot - Company Profits')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

Box Plot:
A box plot, also known as a box-and-whisker plot, is a statistical representation of the distribution of a dataset. It displays the minimum, first quartile, median, third quartile, and maximum of a dataset in a concise manner. The box represents the interquartile range (IQR), and the "whiskers" extend to show the rest of the distribution, with possible outliers plotted as individual points.

Why is it used?
Box plots are used for several reasons:

Summary of Distribution: They provide a concise summary of the distribution of a dataset, indicating the central tendency, spread, and presence of outliers.

Comparison: Box plots are useful for comparing the distributions of different datasets or groups.

Identifying Outliers: They visually highlight potential outliers in the data.

Visualizing Quartiles: The box represents the interquartile range (IQR), allowing a clear visualization of the spread of the central portion of the data.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot - Comparison of Box 1 and Box 2')
plt.show()
```
