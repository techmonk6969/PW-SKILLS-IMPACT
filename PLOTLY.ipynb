Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.

```python
pip install seaborn plotly
```

```python
import seaborn as sns
import plotly.express as px

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Create a scatter plot with Plotly Express
scatter_plot = px.scatter(titanic_data, x='age', y='fare', title='Scatter Plot of Age vs Fare in Titanic',
                          labels={'age': 'Age', 'fare': 'Fare'})

# Show the plot
scatter_plot.show()
```

Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python
import plotly.express as px

# Load the tips dataset from Plotly Express
tips_data = px.data.tips

# Create a histogram with pattern_shape and color parameters
histogram = px.histogram(tips_data, x='sex', y='total_bill', color='day', pattern_shape='smoker',
                         title='Histogram of Total Bill Amount by Sex with Smoker Pattern and Day Color',
                         labels={'sex': 'Sex', 'total_bill': 'Total Bill'})

# Show the plot
histogram.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.
Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.

```python
import plotly.express as px

# Load the iris dataset from Plotly Express
iris_data = px.data.iris()

# Create a scatter matrix plot
scatter_matrix = px.scatter_matrix(iris_data, dimensions=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
                                   color='species', title='Scatter Matrix Plot of Iris Dataset',
                                   labels={'species': 'Species'})

# Show the plot
scatter_matrix.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

```python
import plotly.express as px

# Example data
data = [1.2, 1.4, 1.7, 1.8, 2.1, 2.2, 2.5, 2.8, 2.9, 3.0]

# Create a distplot with Plotly Express
distplot = px.histogram(data, marginal='box', title='Distplot Example',
                        labels={'value': 'Observation'})

# Show the plot
distplot.show()
```
